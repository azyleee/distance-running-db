```python
import pandas as pd 
import sqlite3

rows_in_memory = 100000 # maximum number of rows that can be kept in memory at a time 

df = pd.read_csv('./raw-data-kaggle.csv', delimiter=';')

df
```

keep the data from each athlete in their own table

```python
conn = sqlite3.connect('distance_running_kaggle.db')

for chunk in pd.read_csv('raw-data-kaggle.csv', chunksize=rows_in_memory, delimiter=';'):
    for athlete, group in chunk.groupby('athlete'):
        table_name = athlete
        group.to_sql(table_name, conn, if_exists='append', index=False)

conn.close()
```

Create a table with everyone in it - SQL queries don't suport querying all tables in a dataset without explicit naming

```python
conn = sqlite3.connect('distance_running_kaggle.db')

df.to_sql('everyone', conn, if_exists='replace', index=False)

conn.close()
```

```python
conn = sqlite3.connect('distance_running_kaggle.db')
cursor = conn.cursor()

with open('queries.sql') as f:
    cursor.execute(f.read())
    print(cursor.fetchone()[0])  # prints the result

conn.close()
```
